```python
import pandas as pd
import numpy as np
```

```python
dataframe=pd.read_csv("FullDataFrame.csv")
```

```python
dataframe.shape
```

```python
dataframe.columns
```

```python
#armamos un listado de las columnas que no nos sirven
bad_col=["Unnamed: 0","Business Name","Company Name","DND","Assigned","Address (full)",'Street address', 'City','Postal Code','Birth Date','Offers','Products',"Country","Industry","Lead Offer","Aux for Power Dialer"]
```

```python
#dropeamos las columnas segun la lista que armammos

dataframe_dropped=dataframe.drop(columns=bad_col)
```

```python
dataframe_dropped.columns
```

```python
#Armemos un subdataframe con los IDS, Ad, Day of week (dice si fue convertido) y SubAccount
df1=dataframe_dropped[["Contact Id","Ad","Day Of Week","SubAccount"]]
```

```python
df1
```

```python
df1.info()
```

```python
#drop los leads no convertidos y sin anuncio

df1.replace('-',np.nan,inplace=True)
df1.dropna(inplace=True)
```

```python
df1.info()
```

```python
#Realizo el conteo de conversions de cada anuncio

resume=df1.value_counts(subset=['Ad','SubAccount']).to_frame()
```

```python
#El nombre del anuncio y la subcuenta quedaron como indices, entonces los convierto a columnas
resume.reset_index(['Ad','SubAccount'],inplace=True)

#Cambio el nombre de la columna que tiene el conteo de las conversions
resume.rename(columns={0:'Conversions'},inplace=True)
```

```python
resume
```

```python
# resume.to_csv("Convs por Ads.csv") 
```

# Ahora hacemos el mismo analisis pero agregando el conversion rate

```python
df2=dataframe_dropped[["Contact Id","Ad","Day Of Week","SubAccount"]]
```

```python
df2.Ad.replace('-',np.nan,inplace=True)
df2.dropna(subset='Ad',axis=0,how='any',inplace=True)
```

```python
df2.info()
```

```python
resume2=df2.value_counts(subset=['Ad','SubAccount']).to_frame()

#El nombre del anuncio y la subcuenta quedaron como indices, entonces los convierto a columnas
resume2.reset_index(['Ad','SubAccount'],inplace=True)

#Cambio el nombre de la columna que tiene el conteo de las conversions
resume2.rename(columns={0:'TotalLeads'},inplace=True)
```

```python
resume2
```

```python
resume3=resume2.merge(resume,how='outer',on=['Ad','SubAccount'])
```

```python
resume3.replace(np.nan,0, inplace=True)
```

```python
resume3
```

```python
resume3.dtypes
```

```python
resume3.Conversions=resume3.Conversions.astype('int64')
```

```python
resume3
```

```python
resume3['ConversionRate']=(resume3.Conversions/resume3.TotalLeads)
```

```python
resume3
```
